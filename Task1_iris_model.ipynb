```python
# Import necessary libraries
import pandas as pd
from sklearn.datasets import load_iris
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load Iris dataset
column = ['Sepal Lenght' , 'Sepal Width' , 'Petal Length' , 'Petal Width', 'Class_Label']
df = pd.read_csv('iris.data', names=column)
```

```python
# Display the first few rows of the dataset
print(df.head(50))
```

```python
# Check the basic statistics of the data
print(df.describe())
```

```python
df.info()
```

```python
# Visualize pairwise relationships using seaborn
sns.pairplot(df, hue='Class_Label')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
import pickle

# Input and output data
X = df.drop(columns=['Class_Label'])
Y = df['Class_Label']

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.30)
```

```python
#logidtic regression
model = LogisticRegression(max_iter=500)
model.fit(x_train.values, y_train)
print("Logistic Regression Accuracy:", model.score(x_test, y_test)*100)
```

```python
#model training 
model.fit(x_train.values, y_train.values)
```

```python
#K-nearest neighbors
model = KNeighborsClassifier()
model.fit(x_train.values, y_train.values)
pridict = model.score(x_test, y_test)*100
print("K-nearest neighbors Accuracy:", pridict)
```
