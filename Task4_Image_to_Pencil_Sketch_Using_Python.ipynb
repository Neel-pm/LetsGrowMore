```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('MainAfter.jpg')
plt.imshow(img)
plt.title('original')
plt.show()
```

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image)
plt.title("gray image")
plt.show()
```

```python
inv=255-gray_image
plt.title('Inverted imege')
plt.imshow(inv,cmap='gray')
plt.show()
```

```python
blur=cv2.GaussianBlur(inv,(21,21),sigmaX=0,sigmaY=0)
plt.title('blur applied')
plt.imshow(blur)
plt.show()
```

```python
inverted_blurred = 255 - blur
plt.title('inverted blur image')
plt.imshow(inverted_blurred)
plt.show()
```

```python
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
plt.title('pencil sketch image')
plt.imshow(pencil_sketch)
plt.show()
```
